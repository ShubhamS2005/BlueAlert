```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample GDP data
data = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020, 2021]*2,
    'GDP': [2.5, 2.7, 2.9, 2.3, 2.0, 1.8, 2.1, 3.0, 3.2, 3.4, 3.6, 3.8, 4.0, 4.2],
    'Period': ['Recession']*7 + ['Non-Recession']*7
}
df = pd.DataFrame(data)

fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Plot recession GDP
recession = df[df['Period'] == 'Recession']
axes[0].plot(recession['Year'], recession['GDP'], marker='o', color='red')
axes[0].set_title('GDP During Recession Period')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('GDP Growth (%)')
axes[0].grid(True)

# Plot non-recession GDP
non_recession = df[df['Period'] == 'Non-Recession']
axes[1].plot(non_recession['Year'], non_recession['GDP'], marker='o', color='green')
axes[1].set_title('GDP During Non-Recession Period')
axes[1].set_xlabel('Year')
axes[1].grid(True)

plt.suptitle('GDP Variations During Recession vs Non-Recession Periods')
plt.savefig('Subplot.png')
plt.show()
```

```python
from jupyter_dash import JupyterDash
from dash import html, dcc, Input, Output
import plotly.express as px
import pandas as pd

# Sample data (15 records, consistent across 3 vehicle types and 5 years)
data = {
    'Year': [2017, 2018, 2019, 2020, 2021] * 3,
    'Vehicle_Type': ['Sedan'] * 5 + ['SUV'] * 5 + ['Truck'] * 5,
    'Sales': [12000, 11500, 9000, 6000, 8000,
              13000, 12500, 10000, 7000, 8500,
              11000, 10800, 8500, 5000, 7000],
    'Revenue': [240000, 230000, 180000, 120000, 160000,
                260000, 250000, 200000, 140000, 170000,
                220000, 215000, 170000, 100000, 150000]
}
df = pd.DataFrame(data)

# Initialize Dash app
app = JupyterDash(__name__)
app.title = "Automobile Recession Report Dashboard"

# Layout
app.layout = html.Div([
    html.H1(
        "Automobile Recession Report Dashboard",
        style={
            'textAlign': 'center',
            'fontSize': '36px',
            'color': '#222',
            'padding': '20px'
        }
    ),

    # Dropdowns
    html.Div([
        html.Label("Select Vehicle Type:", style={'fontSize': '20px'}),
        dcc.Dropdown(
            id='vehicle-dropdown',
            options=[{'label': vt, 'value': vt} for vt in df['Vehicle_Type'].unique()],
            value='Sedan',
            clearable=False
        )
    ], style={'width': '50%', 'margin': 'auto', 'marginBottom': '30px'}),

    html.Div([
        html.Label("Select Statistic:", style={'fontSize': '20px'}),
        dcc.Dropdown(
            id='statistic-dropdown',
            options=[
                {'label': 'Sales', 'value': 'Sales'},
                {'label': 'Revenue', 'value': 'Revenue'}
            ],
            value='Sales',
            clearable=False
        )
    ], style={'width': '50%', 'margin': 'auto', 'marginBottom': '30px'}),

    # Output display
    html.Div(
        id='output-container',
        className='output-div',
        style={
            'textAlign': 'center',
            'fontSize': '18px',
            'color': '#444',
            'padding': '20px',
            'border': '1px solid #ccc',
            'borderRadius': '8px',
            'width': '60%',
            'margin': 'auto',
            'marginTop': '20px',
            'backgroundColor': '#f9f9f9'
        }
    ),

    # Recession Graph (Task 2.5)
    dcc.Graph(id='recession-report-graph', style={'marginTop': '30px'}),

    # Yearly Summary Graph (Task 2.6)
    dcc.Graph(id='yearly-report-graph', style={'marginTop': '30px'})
],
style={'backgroundColor': 'white', 'padding': '50px'}
)

# Callback for both graphs and text
@app.callback(
    [Output('output-container', 'children'),
     Output('recession-report-graph', 'figure'),
     Output('yearly-report-graph', 'figure')],
    [Input('vehicle-dropdown', 'value'),
     Input('statistic-dropdown', 'value')]
)
def update_dashboard(vehicle, stat):
    # Filter by vehicle type
    filtered_df = df[df['Vehicle_Type'] == vehicle]

    # Line graph (recession-style trend for selected vehicle type)
    fig1 = px.line(
        filtered_df,
        x='Year',
        y=stat,
        markers=True,
        title=f"{stat} Trend for {vehicle} (Recession View)",
        template='plotly_white'
    )

    # Bar chart: Yearly total of stat across all vehicle types
    yearly_total = df.groupby('Year')[stat].sum().reset_index()
    fig2 = px.bar(
        yearly_total,
        x='Year',
        y=stat,
        title=f"Total {stat} Across All Vehicle Types by Year",
        text_auto=True,
        template='plotly_white'
    )

    output_text = f"You selected Vehicle Type: {vehicle} and Statistic: {stat}."
    return output_text, fig1, fig2

# Run app in Jupyter
app.run(mode='inline')
```
